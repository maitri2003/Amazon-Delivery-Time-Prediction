```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/amazon_delivery_clean.csv")

sns.histplot(df['Delivery_Time'], bins=30, kde=True)
plt.title("Distribution of Delivery Times (hours)")
plt.show()
```

```python
sns.scatterplot(data=df, x='distance_km', y='Delivery_Time', hue='Traffic')
plt.title("Distance vs Delivery Time by Traffic Level")
plt.show()
```

```python
sns.boxplot(data=df, x='Traffic', y='Delivery_Time')
plt.title("Delivery Time by Traffic Condition")
plt.show()
```

```python
sns.boxplot(data=df, x='Weather', y='Delivery_Time')
plt.title("Delivery Time by Weather Condition")
plt.show()
```

```python
sns.scatterplot(data=df, x='Agent_Rating', y='Delivery_Time')
plt.title("Agent Rating vs Delivery Time")
plt.show()
```

```python
sns.boxplot(data=df, x='Agent_Age', y='Delivery_Time')
```

```python
df.groupby("Category")['Delivery_Time'].mean().sort_values().plot(kind='bar')
plt.title("Average Delivery Time by Product Category")
plt.ylabel("Avg Hours")
plt.show()
```

```python
sns.boxplot(data=df, x='Area', y='Delivery_Time')
```

```python
df['Order_Datetime'] = pd.to_datetime(df['Order_Datetime'], errors='coerce')

# Now extract features
df['Order_Hour'] = df['Order_Datetime'].dt.hour
df['Order_Weekday'] = df['Order_Datetime'].dt.day_name()

```

```python
sns.boxplot(data=df, x='Order_Weekday', y='Delivery_Time')
plt.title("Delivery Time by Day of Week")
plt.show()

sns.lineplot(data=df, x='Order_Hour', y='Delivery_Time')
plt.title("Delivery Time Trend by Hour")
plt.show()
```

```python
corr = df[['Delivery_Time','distance_km','Agent_Age','Agent_Rating']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
